```python
## codigo consolidado

#Libs

import pandas as pd
from unidecode import unidecode

# Funcoes
def _norm_local(s: str) -> str:
    """Normaliza: minúsculo, sem acento, sem espaços extras."""
    return unidecode((str(s) or "").strip().lower())


arquivo = "data/receitas_brutas.xlsx"

xls = pd.ExcelFile(arquivo)

print('Lista de TODAS as planilhas:', xls.sheet_names)

lista_sheets = xls.sheet_names

lista_sheets = lista_sheets[1:len(lista_sheets)-1]

print('Lista de planilhas úteis:', lista_sheets)

print('Iniciando leitura da planilha...')

df_final = pd.DataFrame()

for sheet in lista_sheets:
    print(f'{sheet} - ⚡Start')

    df_sheet = pd.read_excel(arquivo, sheet_name=sheet,header=5)

    df_sheet = df_sheet.dropna(subset=['INGREDIENTE'])

    df_sheet.reset_index(inplace=True)

    linha_do_total = df_sheet[df_sheet['INGREDIENTE'] == 'Total'].index[0]

    df_sheet = df_sheet[:linha_do_total]

    df_sheet = df_sheet[['Unnamed: 0', 'INGREDIENTE', 'QUANTIDADE USADA NA RECEITA', 'UN']]

    df_sheet['Unnamed: 0'] = sheet

    df_sheet = df_sheet.rename(columns={'Unnamed: 0': 'prato'})

    df_final = pd.concat([df_final, df_sheet], ignore_index=True)

    print(f'{sheet} - ✅ Done \n')

df_final_tratado = df_final.copy()
df_final_tratado['INGREDIENTE'] = (
    df_final_tratado['INGREDIENTE']
        .astype(str)
        .str.strip()
        .str.lower()
        .apply(unidecode)
)

df_final_tratado['prato'] = (
    df_final_tratado['prato']
        .astype(str)
        .str.strip()
        .str.lower()
        .apply(unidecode)
)

df_final_tratado['UN'] = df_final_tratado['UN'].apply(lambda x: 'un' if x == 'um' else x)

quantidade_pratos = df_final_tratado['prato'].nunique()

print(f'Quantidade de pratos únicos: {quantidade_pratos}')
    
```

```python
df_final_tratado.to_excel('data/receitas_tratadas.xlsx', index=False)
```
